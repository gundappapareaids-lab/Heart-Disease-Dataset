Load Data (Code Cell)

```python
df = pd.read_csv('Heart Disease Dataset.csv')
df.head()
```

Basic Info (Code Cell)

```python
df.shape
df.dtypes
```

Missing Values (Code Cell)

```python
df.isnull().sum()
```

Handle missing:

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

Remove Duplicates (Code Cell)

```python
df.drop_duplicates(inplace=True)
```

Statistics (Code Cell)

```python
df.describe()
```

Manual calculation:

```python
col = df.select_dtypes(include=np.number).columns[0]

print("Mean:", df[col].mean())
print("Median:", df[col].median())
print("Mode:", df[col].mode()[0])
print("Std Dev:", df[col].std())
print("Variance:", df[col].var())
print("Range:", df[col].max() - df[col].min())
print("Midrange:", (df[col].max() + df[col].min())/2)
```

Visualization

```python
plt.hist(df[col])
plt.title("Histogram")
plt.xlabel(col)
plt.ylabel("Frequency")
plt.show()
```

Markdown:

The histogram shows the distribution of the data.

Bar Chart

```python
cat_col = 'sex' # Selected 'sex' as an example categorical column, as there are no 'object' type columns.

sns.countplot(x=df[cat_col])
plt.title(f"Count Plot of {cat_col}")
plt.xlabel(cat_col)
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Boxplot

```python
sns.boxplot(x=df[col])
plt.title("Boxplot")
plt.show()
```

Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Machine Learning (Simple Prediction)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Encoding
df = pd.get_dummies(df, drop_first=True)

X = df.drop(df.columns[-1], axis=1)
y = df[df.columns[-1]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```
